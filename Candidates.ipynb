```python
#from rdflib import Graph
import os
import glob
import utils
response_dir='HEL/files'
os.chdir(response_dir)
```

```python
#f=open(response_dir + '/res1.rdf', 'r')
allData=[]
for resFile in glob.glob('res*.rdf'):
    nilMentions=utils.NILmentions(resFile)
    nilsWithNoCandidates=utils.NILmentionsNoCands(resFile)
    #print("NILs with no candidates", nilsWithCandidates)
    numCands, numMentions = utils.countCandidates(resFile)
    numGoldInCands = utils.recallOfCandidates(resFile)
    recallGoldCands = numGoldInCands/(numMentions-nilMentions)
#    print(numMentions, numCands, numGoldInCands)
    allData.append([resFile, numMentions, numMentions-nilMentions, numGoldInCands, recallGoldCands, numCands, nilMentions, nilsWithNoCandidates])
    #break
```

```python
import pandas
headers = ['filename', 'numMentions', 'nonNilMentions', 'numGoldInCands', 'recallGoldCands', 'numCands', 'nilMentions', 'nilsWithNoCands']
df = pandas.DataFrame(allData, columns=headers)
df
```

```python
print ("Recall of candidates", df["numGoldInCands"].sum()/df["nonNilMentions"].sum())
```

```python
print("Average number of candidates per mention ", df["numCands"].sum()/df["numMentions"].sum())
```
